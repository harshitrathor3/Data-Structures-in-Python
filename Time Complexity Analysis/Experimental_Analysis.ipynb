```python
def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        m = i
        # t=0
        for j in range(i+1, n):
            if arr[j]<arr[m]:
                m = j
        arr[i], arr[m] = arr[m], arr[i]
```

```python
def merge1(ar1, ar2):
    i=0
    j=0
    ans=[]
    while i<len(ar1) and j<len(ar2):
        if ar1[i]<ar2[j]:
            ans.append(ar1[i])
            i+=1
        else:
            ans.append(ar2[j])
            j+=1
    ans+=ar1[i:]
    ans+=ar2[j:]
    return ans
```

```python
def sort1(arr):
    if len(arr)<2:
        return arr
    mid = len(arr)//2
    # a1 = arr[:mid]
    # a2 = arr[mid:]
    a1 = sort1(arr[:mid])
    a2 = sort1(arr[mid:])
    return merge1(a1, a2)
```

```python
def fun(n):
    return list(range(n, 0, -1))
import time
```

```python
n = 1000000
a = fun(n)
# print(a)
start = time.time()
a = sort1(a)
end = time.time()
# print(a)
print(n, end-start)
```

```python
a = fun(n)
# print(a)
start = time.time()
selectionSort(a)
end = time.time()
# print(a)
print(n, end-start)
```
